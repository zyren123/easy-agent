The module contains both a torch implementation and a non-framework reference implementation.

```python
from linear_layer import (
    build_demo_reference_layer,
    build_demo_torch_layer,
    compare_demo_outputs,
    tensor_from_vector,
)

reference_layer = build_demo_reference_layer()
torch_layer = build_demo_torch_layer()
inputs = (2.0, -1.0, 0.5)

reference_layer
```

## What a Linear Layer Does

A linear layer takes an input vector `x` and computes:

`y = Wx + b`

- `W` is the weight matrix
- `b` is the bias vector
- each row of `W` creates one output neuron

```python
reference_layer.input_dim, reference_layer.output_dim
```

## Reference Forward Pass

The non-framework path keeps the arithmetic fully visible.

```python
reference_layer.forward(inputs)
```

## Torch Forward Pass

The torch path should produce the same result for the same weights, bias, and input.

```python
torch_layer(tensor_from_vector(inputs))
```

## Compare Both Paths

This parity check is the key contract for later modules.

```python
compare_demo_outputs()
```

## Per-Neuron Explanation

This is the part beginners usually need to see explicitly: every output neuron uses the whole input vector, but with a different row of weights.

```python
for line in reference_layer.explain_forward(inputs):
    print(line)
```
